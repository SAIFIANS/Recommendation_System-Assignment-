```
# Import the libraries
import numpy as np
import pandas as pd
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
```

```
# Read the dataset
book = pd.read_csv('book.csv', encoding='latin-1')
book
```

```
#Dropping the Unnamed Column
book2 = book.iloc[:,1:]
book2
```

```
#Unique Count of Users
len(book2['User.ID'].unique())
```

```
# Unique Count of Book Titles
len(book2['Book.Title'].unique())
```

```
#Sorting the data by the User ID
book2.sort_values(['User.ID'])
book2
```

```
# Converting the data into a pivot table with User Id as Row number, Book Titles as Columns, Book Ratings as the values in them
book_pivot = book2.pivot_table(index = 'User.ID', columns='Book.Title', values='Book.Rating').reset_index(drop= True)
book_pivot
```

```
#Replacing the Index with the User Id
book_pivot.index = book2['User.ID'].unique()
book_pivot
```

```
#Replacing the Nan with 0
book_pivot.fillna(0, inplace =True)
book_pivot
```

```
#Calculating the cosine similarity
user_sim = 1 - pairwise_distances(book_pivot.values, metric= 'cosine')
user_sim
```

```
# Converting it to a dataframe
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```
# Replacing the index and columns with User ID
user_sim_df.index = book2['User.ID'].unique()
user_sim_df.columns = book2['User.ID'].unique()
user_sim_df
```

```
# Nullifying the diagonal Values
np.fill_diagonal(user_sim, 0)
user_sim_df
```

```
# Most Similar Users
user_sim_df.idxmax(axis=1)
```
